## Read & Display the Image // Tryout

```python
import cv2
from PIL import Image
```

### Reading the image

```python
_path = "./data/Road_in_Norway_Wikipedia.jpg"
img = cv2.imread(_path)
```

### Displaying the image

```python
# cv2.imshow("display_name_for_the_image", image)
cv2.imshow("Image", img)

# This operation waits for the user's reponse
# and this will prevent the kernel from crashing
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### 'matplotlib' can also be used to perform the same operation

```python
import matplotlib.pyplot as plt   ## Importing the library
img_read = plt.imread(_path)      ## To read the image
plt.imshow(img_read)              ## To display the image
plt.show()
```

### Copying the image using cv2

```python
img_copy = img.copy()
plt.imshow(img_copy)
```

### However, there are some pointers to remember
***
<div class="alert alert-block alert-info">
<font color = 'black'>We used cv2 to read the image, but used <b>'plt'</b> to display the same. Since, they operate on different combination of color channels, we are being subjected to such inference. <b>Here, 'cv2' is reading the image in 'BGR' color format whereas we are expecting it to be under 'RGB' which is what we are expecting.</b>
</font></div>
       
* If we convert the resulting read image into **'RGB'**, we can understand better. We can do it using **'cv2.COLOR_BGR2RGB'** in cv2.


* The copied image using **.copy()** method will having different address as compared to the one it's copied from. We can find out by using id(image_name_here) function.


* We can get rid of the above displayed statement "<matplotlib.image.AxesImage at #############>" by simply including plt.show() in the next.

### We will now see the result after implementing the same below. Let's see.
Converting from BGR to RGB and displaying the output.

```python
img_copy_cvt = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
plt.imshow(img_copy_cvt)
```

Getting rid of the message: **<matplotlib.image.AxesImage at 0x238d8820c10>** using **plt.show()**

```python
img_copy_cvt = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
plt.imshow(img_copy_cvt)
plt.show()
```

Accessing the **memory address** of the image using **id()**

```python
print(f'img: {id(img)}\nimg_copy: {id(img_copy)}\nimg_copy_cvt: {id(img_copy_cvt)}')
```

<div class="alert alert-block alert-warning">
<font color = 'black'>If we do img_dup = img the any changes being made on img will be impacting the img_dup as it is accessing the <b>same memory address</b></font></div>

```python
img_dup = img
img_dup_copy = img.copy()
img[:,:,1] = 0 # Setting red channel to zero
plt.figure(figsize = (10,10))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1, 3, 2)
plt.title("Dupliate Image")
plt.imshow(img_dup)

plt.subplot(1, 3, 3)
plt.title("Dupliate Copied Image")
plt.imshow(img_dup_copy)

plt.show()
```

**Why is that?**
Let's see their memory address once.

```python
print(f'Memory Adrress\nimg_dup: {id(img_dup)}\nimg: {id(img)}\nimg_dup_copy: {id(img_dup_copy)}')
```

Now we know why it happened in the first place and we also know that why the third image **Duplicate Copied Image** from left is coming in such format. Yes! We'd need to convert to **RGB** format to see the expected.

```python
plt.imshow(cv2.cvtColor(img_dup_copy, cv2.COLOR_BGR2RGB))
plt.show()
```
