### Importing the libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Reading the image

```python
path = "./data/Road_in_Norway_Wikipedia.jpg"
img_p = plt.imread(path)
plt.imshow(img_p)
plt.show()
```

### Plotting **Histogram** for the provided image
<p>For histogram, we will be using cv2(OpenCV) to plot the histogram</p>

**cv2.calcHist()** - cv2.calcHist(images, channels, mask, histSize, ranges)

```python
img_p_gray = cv2.cvtColor(img_p, cv2.COLOR_BGR2GRAY)

def fetch_y_vals(hist_vals):
    y_val = []
    for h in hist_vals:
        for value in h:
            y_val.append(value)
    return y_val

hist_vals = cv2.calcHist([img_p_gray], [0], None, [256], [0, 255])
y_value = fetch_y_vals(hist_vals)        
x_value = [x for x in range(256)]

plt.figure(figsize = (15,15))
plt.subplot(2, 1, 1)
plt.title("Original Grayscaled Image")
plt.imshow(img_p_gray, cmap = "gray")

plt.subplot(2, 1, 2)
plt.bar(x_value,y_value)

plt.show()
img_p_gray.dtype
```

### Let's perform some Intensity transformations

```python
pixel_block = np.array([0, 1, 1, 0.5, 0.5, 0.5, 0.5, 0.5, 1]).reshape(3,3)
pixel_block*=255.0
print(pixel_block)
plt.imshow(pixel_block, cmap = "gray")
plt.show()
```

Increasing the **intensity** of each pixel

```python
def intensity_transform(img):
#     row, col = img.shape
    for row in range(img.shape[0]):
        for col in range(img.shape[1]):
            img[row, col] = img[row, col]*2 + 1 
    return img

img_intense = intensity_transform(pixel_block)
img_intense = img_intense.astype('int8')
print(img_intense)

plt.imshow(img_intense, cmap = "gray")
plt.show()
```

### Working with Image Negatives

```python
img_c = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
plt.imshow(img_c, cmap = "gray")

# Inverting the pixel values

def apply_negative(img_name_here):
    for r in range(img_name_here.shape[0]):
        for c in range(img_name_here.shape[1]):
            img_name_here[r, c] = -img_name_here[r, c] + 255
    return img_name_here

fetch_neg = apply_negative(img_c)
plt.imshow(fetch_neg, cmap = "gray")
plt.show()
```

### Comparing the histograms between the two image <i>(Original v/s Negative)</i>

```python
hist_neg = cv2.calcHist([fetch_neg], [0], None, [256], [0, 255])
y_val_neg = fetch_y_vals(hist_neg)

plt.figure(figsize = (10, 10))

plt.subplot(1,2,1)
plt.title("Original")
plt.bar(x_value, y_value) # x and y values of the Original image

plt.subplot(1,2,2)
plt.title("Inverted")
plt.bar(x_value, y_val_neg) # x and y values of the Negative/Inverted image

plt.show()
```

### Performing linear transformations like <u>Brightness</u> and <u>Contrast</u>

We can use this linear model where <font color="blue"> $ {\alpha}{f[i,j]} + {\beta} $ </font>
where **${\alpha}$** represents simple contrast control and **${\beta}$**
represents simple brightness control. Let's perform this transformation.

```python
## Defining the linear transformation function
def linear_transform(img_here, a, b):
    img_copy = np.copy(img_here)
    for row in range(img_copy.shape[0]):
        for col in range(img_copy.shape[1]):
            img_copy[row, col] = a*img_here[row, col] + b
            
    return img_copy

img_transform = cv2.imread(path, cv2.IMREAD_GRAYSCALE)

_alpha = 1  # contrast control
_beta = 100 # brightness control

apply_image = linear_transform(img_transform, _alpha, _beta)
plt.imshow(apply_image, cmap = "gray")
plt.show()
```

Using in-built function **cv2.convertScaleAbs()** to perform the transformation. Let see the result.

```python
image_transformed = cv2.convertScaleAbs(img_transform, alpha = _alpha, beta = _beta)
plt.imshow(image_transformed, cmap = "gray")
plt.show()
```

Let's compare the histograms between the original image and the intensified one.

```python
hist_intense = cv2.calcHist(image_transformed, [0], None, [256], [0, 255])
y_val_intensified = fetch_y_vals(hist_intense)

plt.figure(figsize = (15, 15))
plt.subplot(1, 2, 1)
plt.title("Original")
plt.bar(x_value, y_value) # x and y values of the Original image

plt.subplot(1, 2, 2)
plt.title("Intensified")
plt.bar(x_value, y_val_intensified) # x and y values of the intensified image
plt.show()
```

Using **optimization** techniques to make it more **well-distributed**, which is why we have histogram equalization.
### What is **Histogram Equalization**?
The algorithm that uses images' histogram to adjust the contrast using **<font color = "Blue">cv2.equalizeHist(image_name_here)</font>**

```python
new_equalize = cv2.equalizeHist(img_transform)

plt.figure(figsize = (15, 15))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_transform, cmap = "gray")

plt.subplot(1,2,2)
plt.title("Equalized Image")
plt.imshow(new_equalize, cmap = "gray")

plt.show()
```

### Thresholding and simple segmentation

**Thresholding** allows to **apply threshold to every pixel** which can be used **extract an object** from an image which is known as **segmentation**.

Defining the threshold function

```python
def apply_threshold(image_name_here, threshold_val, min_val, max_val):
    height, width = image_name_here.shape
    final_img = np.zeros((height, width), dtype = np.uint8)
    
    for r in range(height):
        for c in range(width):
            if image_name_here[r, c] >= threshold_val:
                final_img[r, c] = max_val
            else:
                final_img[r, c] = min_val
    return final_img
```

Let's use the custom image of shape **(3,3)** dimension we created of name **pixel_block**

```python
pixel_block = pixel_block/255.0
```

```python
threshold_val = 1
min_val = 0
max_val = 2

pixel_block = pixel_block.astype(np.uint8)

pixel_block_thresh = apply_threshold(pixel_block, threshold_val, min_val, max_val)
plt.imshow(pixel_block_thresh, cmap = "gray")
plt.show()
```

#### Let's use a different image here.

```python
path_2 = "./data/autonomous_car.png"
new_img = cv2.imread(path_2, cv2.IMREAD_GRAYSCALE)
plt.imshow(new_img, cmap="gray")
plt.show()
```

```python
threshold_val = 80
min_val = 0
max_val = 255
new_img_thresh = apply_threshold(new_img, threshold_val, min_val, max_val)

hist_new = cv2.calcHist([new_img_thresh], [0], None, [256], [0, 255])
y_val_new = fetch_y_vals(hist_new)

plt.figure(figsize = (15,15))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(new_img, cmap = "gray")

plt.subplot(2, 2, 2)
plt.title("Image after thresholding")
plt.imshow(new_img_thresh, cmap = "gray")

plt.subplot(2, 2, 3)
plt.title("Original Histogram")
plt.bar(x_value, y_value)

plt.subplot(2, 2, 4)
plt.title("Histogram after thresholding")
plt.bar(x_value, y_val_new)

plt.tight_layout()
plt.show()
```

### Let's looking into the inbuild threshold method provided by OpenCV - <font color = "Blue">cv2.threshold()</font>
cv2.threshold() **takes in three parameters**.
* image_name
* Threshold value
* Maximum pixel intensity value
* Threshold type

and **returns two output values**,
* Threshold
* Thresholded image

**cv2.threshold(image_name_here, threshold_value, max_intensity_value, threshold_type)**

```python
_threshold = 80
max_intensity_value = 255

# cv2.THRESH_BINARY, here output will either be 0 or 255
ret, thresholded_img = cv2.threshold(new_img, _threshold, max_intensity_value, cv2.THRESH_BINARY)

plt.imshow(thresholded_img, cmap = "gray")
plt.show()
```

### But, there's a catch.
---

<div class="alert alert-block alert-info"><font color = "Black">

Often it's hard to determine the threshold values for us to extract the required object a given environment, and is most likely due to surrounding adjacent colors in the mapped environment which **makes it's more challenging to distinguish** our **region of interest**. 

**cv2.THRESH_OTSU** the threshold type which will help us in selecting the threshold value automatically. Let see how it's impacting our image.

---
For applying the threshold type **cv2.THRESH_OTSU**, we'd have to provide four parameters including the threshold type.
**<p>cv2.threshold(image_name_here, min_pixel_intensity_value, max_pixel_intensity_value, cv2.THRESH_OTSU)</p>**
</font></div>

```python
ret, thresh_img_otsu = cv2.threshold(new_img, 0, 255, cv2.THRESH_OTSU)
plt.imshow(thresh_img_otsu, cmap = "gray")
plt.show()
```

```python
hist_otsu = cv2.calcHist([thresh_img_otsu], [0], None, [256], [0, 255])
y_val_new = fetch_y_vals(hist_otsu)

plt.figure(figsize = (15,15))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(new_img, cmap = "gray")

plt.subplot(2, 2, 2)
plt.title("Image after applying cv2.THRESH_OTSU")
plt.imshow(thresh_img_otsu, cmap = "gray")

plt.subplot(2, 2, 3)
plt.title("Original Histogram")
plt.bar(x_value, y_value)


plt.subplot(2, 2, 4)
plt.title("Histogram after applying cv2.THRESH_OTSU")
plt.bar(x_value, y_val_new)

plt.tight_layout()
plt.show()
```
