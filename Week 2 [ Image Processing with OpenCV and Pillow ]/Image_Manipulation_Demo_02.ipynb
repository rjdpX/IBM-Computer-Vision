## Flip the Image
---

#### First, we import the required libraries

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

#### Reading the image

```python
_path = "./data/Road_in_Norway_Wikipedia.jpg"
img = plt.imread(_path)
img_cvt = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

#### Displaying the image

```python
plt.imshow(img)
plt.show()
```

<div class="alert alert-block alert-info">
<font color = 'black'> We are <b>not</b> required to convert from 'BGR' to 'RGB' here if using <b>plt.imread()</b>
    to read the image</font>
</div>

### Now, we'll be peforming some operations on the image

We can flip an image by changing the index value of the pixel or intensity of the image. We'll using **PIL** library o perform the opertaion.

```python
from PIL import ImageOps
```

```python
img_p = Image.open(_path)
img_p.show()
```

ImageOps() flips the image w.r.t x axis or flips the y axis **top_to_bottom**

```python
img_flip_p = ImageOps.flip(img_p)
plt.figure(figsize = (10,10))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img_p)

plt.subplot(1, 2, 2)
plt.title('Flipped Image')
plt.imshow(img_flip_p)
plt.show()
```

#### We can also mirror the image using ImageOps.mirror()

```python
img_mirror_p = ImageOps.mirror(img_p)
plt.title("Mirrored Image")
plt.imshow(img_mirror_p)
plt.show()
```

<div class="alert alert-block alert-info">
    <font color = "black">
    <b>cmap is ignored when X has <font color="red">RGB(A)</font> information.</b>
    However, if the image were an array of shape (M,N) , then the cmap controls the colormap used to display the values. We'll see the same below.
    </font>
</div>

```python
img_gray_p = ImageOps.grayscale(img_p)

## ImageOps.posterize(image_name_here, number_of_bits_1_to_8_to_keep_for_each_channel)
img_post_p = ImageOps.posterize(img_p, 2)

plt.figure(figsize = (15,15))
plt.subplot(1,2,1)
plt.title("Grayscaled Image")
plt.imshow(img_gray_p)

plt.subplot(1,2,2)
plt.title("Posterized Image")
plt.imshow(img_post_p)
plt.show()
```

### Using colormap (cmap):
---
<div><font size = 3 color = 'blue'><b>plt.imshow(image_name_here, cmap = 'gray')</b></font></div>

```python
plt.title("Grascaled Image cmapped")
plt.imshow(img_gray_p, cmap = 'gray')
plt.show()
```

The above operation is only being applied to grayscaled image, if used on **RGB** image, it won't affecting the same. We'll see that here below.

```python
plt.imshow(img_p, cmap = 'gray')
plt.show()
```

We can also use **.transpose()** method to define specific transformation we are looking for.

```python
img_flip_t_b = img_p.transpose(Image.FLIP_TOP_BOTTOM)
plt.title("Flipped image using .transpose()")
plt.imshow(img_flip_t_b)
plt.show()
```

### We'll now see how to flip image using <font color = "blue">OpenCV</font>
---

```python
img_c = cv2.imread(_path)
cv2.imshow("Image Output", img_c)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#### Converting the image **BGR** to **RGB** to show through plt.imshow()

```python
img_c = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)
plt.imshow(img_c)
plt.show()
```

### Flipping the image using <font color = "blue">cv2.flip()</font>
---
**Understanding flip_code**

* A flag to specify how to flip the array
* 0 means flipping around the x-axis and positive value (for example, 1) means flipping around y-axis.
* Negative value (for example, -1) means flipping around both axes.

```python
img_flip0_c = cv2.flip(img_c, 0)  # cv2.flip(image_name_here, flip_code), here flipping along x-axis
img_flip1_c = cv2.flip(img_c, 1)  # Flipping along the y-axis
img_flip2_c = cv2.flip(img_c, -1) # Flipping along both the axis

plt.figure(figsize = (15,15))

plt.subplot(1,3,1)
plt.title("Flipped by x-axis")
plt.imshow(img_flip0_c)

plt.subplot(1,3,2)
plt.title("Flipped by y-axis")
plt.imshow(img_flip1_c)

plt.subplot(1,3,3)
plt.title("Flipped by both-axes")
plt.imshow(img_flip2_c)

plt.show()
```

### Rotate the image using <font color = "blue">cv2.rotate()</font>

```python
img_rot_c = cv2.rotate(img_c, cv2.ROTATE_90_CLOCKWISE)
plt.imshow(img_rot_c)
plt.show()
```
