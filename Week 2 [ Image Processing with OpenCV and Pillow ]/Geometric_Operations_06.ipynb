### <font color = "Blue">Geometric</font> Transformations
* Scaling
* Translation
* Rotation

In geometric transformaton we change the coordinationates of the image(x, y). We will look into a **subset** of the geometric transformation called **Affine Transformation**.

* What is Affine Transformation?
> Affine Transformation deals with various types of affine transformation provided we are maintaining the images' lines and parallelism nature like scaling, translation, rotation, reflection etc.

Let's **import** the required libraries first!

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

### Reading the image using PIL

```python
path_ = "./data/waymo_jaguar_i_pace_electric_suv.jpg"
img_p = Image.open(path_)

plt.imshow(img_p)
plt.show()

print(img_p.size)
```

### Scaling using PIL library

Inorder to scale, we need two arguments. **image_name.<font color = "Red">resize</font>( (new_height_here, new_width_here) )**
* height
* width

```python
width, height = img_p.size

new_height = 600 # increasing the width
new_width = width # keeping the same width
new_scaled_img = img_p.resize((new_width, new_height))

plt.figure(figsize = (15,15))

plt.subplot(1,2,1)
plt.title("original Image")
plt.imshow(img_p)

plt.subplot(1,2,2)
plt.title("Scaled Image")
plt.imshow(new_scaled_img)

plt.tight_layout()
plt.show()
```

.rotate() helps to rotate the image while the exposed background being set to **0** or Black.

### Geometric transformations using <font color = "blue">OpenCV</font>
___

Reading the image using **cv2.imread()**

```python
img_c = cv2.imread(path_)
img_c = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB) # Converting from BGR to RGB
plt.imshow(img_c)
plt.show()
```

#### Applying transformation

* Scaling
* Translation
* Rotation

**Scaling** is performed using **cv2.resize()** method and the **parameters** that are required to perform the operation:

* **image_name** - Input image name which we want to resize.
* **dsize** - Desired size of the output image.
* **fx** - Scaled factor along horizontal axis.
* **fy** - Scaled factor along vertical axis.

**cv2.<font color = "blue">resize</font>(image_name_here, dsize, fx, fy)**
___

Let's see what this gives us!

```python
img_c_scaled = cv2.resize(img_c, None, fx = 1, fy = 2, interpolation = cv2.INTER_CUBIC)

plt.figure(figsize = (15,15))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_c)

plt.subplot(1,2,2)
plt.title("Scaled Image")
plt.imshow(img_c_scaled)

plt.tight_layout()
plt.show()
```

#### Applying translation

**Translation** is performed using cv2.warpAffine() and it requires **Affine Transformational matrix** to be defined based on which translation operation can be performed. We'd need two values specifically to define the operation <font color = "Red">$t_x$</font> and <font color = "Red">$t_y$</font>.

We'll now see how it actually works.

```python
import numpy as np

# defining Affine transformation matrix

M = np.float32([[1, 0, 75],[0, 1, 25]]) # tx = 75, ty = 25 (will be translated by 75px along x-axis and 25px along y-axis)
print(M)
```

```python
img_c_translate = cv2.warpAffine(img_c, M, (img_c.shape[1], img_c.shape[0]))
plt.imshow(img_c_translate)
plt.show()
```

#### Applying rotation

Inorder to perform rotation, we can get the transformation matrix to rotate the image using **cv2.getRotationMatrix2D()**.

Parameters required to perform cv2.getRotationMatrix2D()

* Center - center of the image which can be obtained using cols//2-1 for height and rows//2-1 for width **(cols//2-1, rows//2-1)**
* Angle - Required angle of rotations (user specific)
* Scale - Scaling factor

```python
img_c.shape # Returns (height, width, no_of_channels)
```

```python
rows, cols, _ = img_c.shape
center = (cols//2-1, rows//2-1)
M_rot = cv2.getRotationMatrix2D(center, 45, 1)

img_c_rot = cv2.warpAffine(img_c, M_rot, (rows, cols))
plt.imshow(img_c_rot)
plt.show()
```

## Performing <font color = "Blue">matrix operations</font>

Let's open the image using PIL Image. We'll first read a grayscale image.

```python
path_gray = "./data/rolls_royce_phantom_coupe.jpg"
img_p_gray = Image.open(path_gray)
plt.imshow(img_p_gray)
plt.show()
```

Its **grayscale**, but **having three channels**. We'll see that.

```python
print(np.array(img_p_gray).shape)
```

We can change from **three channels** to **one channel** using **ImageOps**.

```python
from PIL import ImageOps # Importing ImageOps

img_p_covt = ImageOps.grayscale(img_p_gray) # grayscale conversion using img_p_covt

print(img_p_covt.size) # Image size
img_p_covt = np.array(img_p_covt)
print(img_p_covt.shape) # Here, we are looking at image "array" shape and not image size
```

```python
plt.imshow(img_p_covt, cmap = "gray")
plt.show()
```

Here, we can apply algorithms designed for matrices. Like, **svd (Singular Value Decomposition)** where it extract/decomposes our image marix, img_p_covt(in our case) into a product of three matrices. Let's see how we get.

```python
U, s, V = np.linalg.svd(img_p_covt, full_matrices = True)
print(s.shape)
```

We can convert s into a **diagonal** matrix



```python
S = np.zeros(img_p_covt.shape)
S[:S.shape[0],:S.shape[0]] = np.diag(s)
print(S)
```

```python
plt.figure(figsize = (15,15))

plt.subplot(1,2,1)
plt.title("U")
plt.imshow(U, cmap = "gray")

plt.subplot(1,2,2)
plt.title("V")
plt.imshow(V, cmap = "gray")

plt.tight_layout()
plt.show()
```

```python
B = S.dot(V)
A = U.dot(B)
plt.imshow(A, cmap = "gray")
plt.show()
```

```python
plt.figure(figsize = (10,10))

for i, items in enumerate([3, 5, 10, 50, 100, 500]):
    S_select = S[:, :items]
    V_select = V[:items, :]
    final_select = U.dot(S_select.dot(V_select))
    plt.subplot(2,3,i+1)
    plt.title("No of items: " + str(items))
    plt.imshow(final_select, cmap = "gray")

plt.tight_layout()    
plt.show()
```

Through this we can infer that **100 items would be sufficient** to potray the image.
