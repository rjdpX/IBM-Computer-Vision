## <font color = "Blue">Spatial</font> Operations
___

There are many image enhancement techniques that are based out of spatial operations being performed on **local neighbourhood of the neighbouring pixels**. The opertations includes **convolving a spatial mask** over the image and **replacing the values with some resultant weighted value**. The operations include:
* Linear filtering
* Edge detection
* Linear filtering

Let's dive a little bit deeper on how this works.

First, we import the required libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image, ImageOps, ImageFilter # ImageFilter to use pre-define filters
```

```python
path = "./data/Road_in_Norway_Wikipedia.jpg"
img_p = Image.open(path)
plt.imshow(img_p)
plt.show()
```

Converting this image to **grayscale**

```python
img_p_gray = ImageOps.grayscale(img_p)
plt.imshow(img_p_gray, cmap = "gray")
plt.show()
print(img_p_gray.size)
```

Let's add some noise in the image

```python
noise = np.random.normal(1, 30, np.array(img_p_gray).shape)
img_noise = img_p_gray + noise
plt.imshow(img_noise, cmap = "gray")
plt.show()
```

Adding **Salt and Pepper** noise in the image : def add_salt_pepper()

```python
def add_salt_pepper(image):
    
    image_arr = np.array(image)
    row, col = image_arr.shape
    
    pixel = np.random.randint(300, 10000)
    
    for i in range(pixel):
        y_val = np.random.randint(0, col-1)
        x_val = np.random.randint(0, row-1)
        
        image_arr[x_val, y_val] = 255
        
        y_val = np.random.randint(0, col-1)
        x_val = np.random.randint(0, row-1)
        
        image_arr[x_val, y_val] = 0
    
    pixel = np.random.randint(300, 10000)
    
    for i in range(pixel):
        y_val = np.random.randint(0, col-1)
        x_val = np.random.randint(0, row-1)
        
        image_arr[x_val, y_val] = 255
        
        y_val = np.random.randint(0, col-1)
        x_val = np.random.randint(0, row-1)
        
        image_arr[x_val, y_val] = 0
    
    pixel = np.random.randint(300, 10000)
    
    for i in range(pixel):
        y_val = np.random.randint(0, col-1)
        x_val = np.random.randint(0, row-1)
        
        image_arr[x_val, y_val] = 255
        
        y_val = np.random.randint(0, col-1)
        x_val = np.random.randint(0, row-1)
        
        image_arr[x_val, y_val] = 0
        
    return image_arr
```

```python
img_ar = add_salt_pepper(img_p_gray)
plt.figure(figsize = (10, 10))
plt.title("Image with salt and pepper noise!")
noisy_img = Image.fromarray(img_ar)
plt.imshow(noisy_img, cmap = "gray")
plt.show()
```

Creating a kernel whose each value is (1/36) and convolve this kernel over the image.

```python
kernel = np.ones((5,5)) * 1/36
print(kernel)
```

```python
kernel_filter = ImageFilter.Kernel((5,5), kernel.flatten())
```

```python
# kernel = ImageFilter.Kernel((5,5), kernel_filter.flatten())
filtered_img = noisy_img.filter(kernel_filter)

plt.figure(figsize = (15,15))

plt.subplot(1,2,1)
plt.title("Noisy Image")
plt.imshow(noisy_img, cmap = "gray")

plt.subplot(1,2,2)
plt.title("Filtered Image")
plt.imshow(filtered_img, cmap = "gray")
plt.show()
```

Applying Gaussian Blur on the **noisy image**.
Inorder to apply gaussian blur, we can use .filter() method to incorporate which type of filter we'd like to apply.
**ImageFilter** provides a already pre-defined method called as **GaussianFilter**.

* ImageFilter.GaussianBlur

```python
blurred_noisy_img = noisy_img.filter(ImageFilter.GaussianBlur(3))

plt.figure(figsize = (15,15))

plt.subplot(1,3,1)
plt.title("Original Image")
plt.imshow(img_p_gray, cmap = "gray")

plt.subplot(1,3,2)
plt.title("Noisy Image")
plt.imshow(noisy_img, cmap = "gray")

plt.subplot(1,3,3)
plt.title("Blurred Image")
plt.imshow(blurred_noisy_img, cmap = "gray")

plt.tight_layout()
plt.show()
```

###  Image Sharpening

```python
kernel = np.array([-1, -1, -1, -1, 9, -1, -1, -1, -1]).reshape(3,3) # kernel we'll be using
```

```python
sharpen_filter = ImageFilter.Kernel((3,3), kernel.flatten())
sharpen_img = img_p_gray.filter(sharpen_filter)

plt.figure(figsize = (10, 10))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(img_p_gray, cmap = "gray")

plt.subplot(1,2,2)
plt.title("Sharpened Image")
plt.imshow(sharpen_img, cmap = "gray")

plt.tight_layout()
plt.show()
```

Using **pre-defined** filter **ImageFilter.SHARPEN**

```python
sharpen_img_PIL = img_p_gray.filter(ImageFilter.SHARPEN)
plt.imshow(sharpen_img_PIL, cmap = "gray")
plt.show()
```

```python
plt.figure(figsize = (15, 15))

plt.subplot(3,1,1)
plt.title("Original Image")
plt.imshow(img_p_gray, cmap = "gray")

plt.subplot(3,1,2)
plt.title("Sharpened Image using custom kernel")
plt.imshow(sharpen_img, cmap = "gray")

plt.subplot(3,1,3)
plt.title("Sharpened Image using PIL library")
plt.imshow(sharpen_img_PIL, cmap = "gray")

plt.tight_layout()
plt.show()
```

### Fetching edges

```python
img_path_ = "./data/rolls_royce_silver_ghost.jpg"
img_pil = Image.open(img_path_)
img_pil_gray = ImageOps.grayscale(img_pil)
plt.imshow(img_pil_gray, cmap = "gray")
plt.show()
```

```python
img_pil_edge = img_pil_gray.filter(ImageFilter.FIND_EDGES)
plt.imshow(img_pil_edge, cmap = "gray")
```

```python
img_pil_enhance = img_pil_gray.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(img_pil_enhance, cmap = "gray")
```

```python
plt.figure(figsize = (15, 15))

plt.subplot(3,1,1)
plt.title("Original Image")
plt.imshow(img_pil_gray, cmap = "gray")

plt.subplot(3,1,2)
plt.title("Edge Enhance Image")
plt.imshow(img_pil_enhance, cmap = "gray")

plt.subplot(3,1,3)
plt.title("Edges in the Image")
plt.imshow(img_pil_edge, cmap = "gray")

plt.tight_layout()
plt.show()
```

Applying **Median Blur** over the image.

```python
img_p_gray_med = noisy_img.filter(ImageFilter.MedianFilter)
plt.imshow(img_p_gray_med, cmap = "gray")
```

### Let's compare with the noisy and original image
Using plt.subplot()

```python
plt.figure(figsize = (15, 15))

plt.subplot(3,1,1)
plt.title("Original Image")
plt.imshow(img_p_gray, cmap = "gray")

plt.subplot(3,1,2)
plt.title("Image with noise")
plt.imshow(noisy_img, cmap = "gray")

plt.subplot(3,1,3)
plt.title("De-noising using median blur")
plt.imshow(img_p_gray_med, cmap = "gray")

plt.tight_layout()
plt.show()
```
