## Drawing objects on a image
---

First, importing the required libraries and reading a image.
* Import libraries
* Read the image

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
path = "./data/Road_in_Norway_Wikipedia.jpg"
image_p = plt.imread(path)
plt.imshow(image_p)
plt.show()
```

#### Inorder to draw objects on the image we'll use <font color = "blue">ImageDraw to draw PIL image objects</font>

```python
from PIL import ImageDraw
```

```python
img_pil = Image.open(path) 
img_cpy = img_pil.copy()
img_func = ImageDraw.Draw(im = img_cpy) # Creating an object for the image >>> img_cpy

## We can mention the dimension in this sequence, left, upper, right, lower
## This is set of points that I've selected, feel free to tweak on these values

left  = 200
upper = 100
right = 700
lower = 400

draw_shape = [left, upper, right, lower]
img_func.rectangle(xy = draw_shape, fill = "blue")
plt.imshow(img_cpy)
plt.show()
```

## Writing text on images
---

```python
from PIL import ImageFont

# Using the same img_cpy image here.
txt_fnt = ImageFont.truetype("arial.ttf", 100)
img_func.text(xy = (0,0), text = "blue box", font = txt_fnt, fill = (0,0,0))
plt.imshow(img_cpy)
plt.show()
```

### Use OpenCV functions to perform image manipulations

```python
img_c = cv2.imread(path)
img_c = cv2.cvtColor(img_c, cv2.COLOR_BGR2RGB)
st_pt, end_pt = (left, upper), (right, lower)
cv2.rectangle(img_c, pt1 = st_pt, pt2=end_pt, color = (0,0,255), thickness = 5)
plt.imshow(img_c)
plt.show()
```

### Putting text on an image using OpenCV

We'll be using the same image >>>> **img_c**

```python
cv2.putText(img_c, text="Blue Box using OpenCV", org = (200, 600), color = (0, 0, 255), fontFace = 3, \
            fontScale = 2, thickness = 4)
plt.imshow(img_c)
plt.show()
```
